### Interest Over Time Assessment

#### Import necessary libraries

```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
import datetime
from datetime import datetime, date, time
```

#### Collect Pytrend data for the trend you would like to review. Pytrend is the unofficial API for Google Trends.

```python
pytrend = TrendReq()
pytrend.build_payload(kw_list =[ 'hartis consulting', 'IT Consulting', 'Cloud', 'big data', 'AI'], timeframe='today 12-m', geo = 'GB', cat= 1162)
```

```python
pytrend.interest_over_time()
interest_over_time_df = pytrend.interest_over_time()
print(interest_over_time_df)
```

#### Steps to build simple graph using Seaborn from Pytrend data

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
dx = interest_over_time_df.plot.line(figsize = (9,6), title = "Interest Over Time")
dx.set_xlabel ('Date')
dx.set_ylabel ('Trends Index')
dx.tick_params(axis='both', which='major', labelsize=13)
```
