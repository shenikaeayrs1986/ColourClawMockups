# Creating Basic Graphs in Jupyter Notebook

## Getting Started 

### Step One import all necessary libraries

```python
from IPython.display import HTML
import json
import pandas as pd
import matplotlib.pyplot as plt
```

### Step Two add data source. For this basic example we will use a free Specialty Bike data source from Webdatarocks

```python
dataframe= pd.read_json("https://cdn.webdatarocks.com/data/data.json")
dataframe.head()
```

### Step Three create a basic pie graph, bar graph, scatter plot, and line graph using the Speaclity Bike source

#### Pie Graph

```python
dataframe.groupby(['Country']).sum().plot(kind='pie', y='Price', title= 'Specialty Bike Shop', autopct= '%1.0f%%')
```

```python
df= pd.DataFrame({
    'Country': ['Australia', 'Cananda', 'France', 'Germany', 'United Kingdom'], 
    'Color':['Red', 'Yellow','White', 'Yellow', 'Red'],
    'Quantity': [225,90,855,897,115], 
    'Price': [174,507,242, 102, 126], 
    'Discount': [23,17,37,42,44]
})
df.head()
```

#### Bar Graph

```python
df.plot(kind='bar', x='Quantity', y='Price', title= 'Specialty Bike Price Vs Quantity')
plt.show()
```

#### Scatter Plot

```python
df.plot(kind='scatter', x='Quantity', y='Price', title= 'Specialty Bike Price Vs Quantity')
plt.show()
```

#### Multiple Column Line Graph

```python
ax = plt.gca()
df.plot(kind='line', x='Price', y='Discount', ax=ax)
df.plot(kind='line', x='Price', y= 'Quantity', color= 'Violet', ax=ax)
plt.show()
```

#### Additional Bar Graph Examples

```python
df.groupby('Country')['Quantity'].nunique().plot(kind='bar')
plt.show
```

```python
Country= ['Australia', 'Cananda', 'France', 'Germany', 'United Kingdom']
Quantity= [225,90,855,897,115]
Price= [174,507,242, 102, 126]
plt.bar(Country, Quantity, color ='y')
plt.bar(Country, Price, bottom=Quantity, color='g') 
plt.show()
```
